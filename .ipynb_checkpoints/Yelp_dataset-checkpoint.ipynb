# Modeling and testing

## - What leads to more reviews?
- full bar (+) and price range (+)
- Bars/Night life get more reviews

```python
# -scatter
var = ['business_review_count','business_stars','business_PriceRange']
from pandas.tools.plotting import scatter_matrix
scatter_matrix(business[var], alpha=0.2, figsize=(6, 6), diagonal='hist');
```

```python
# -multiple regression
import statsmodels.formula.api as smf
features = 'business_PriceRange+ C(business_Alcohol)+ C(business_NoiseLevel)+ C(business_HasTV)'
est = smf.ols(formula='business_review_count ~'+features, data=business).fit()
est.summary().tables[1]
```

```python
# -standardize the restaurant types
ctg = set([i for ctgs in business.business_categories.str.split(', ') for i in ctgs])
ctg = set(i.replace(']','').replace('[','') for i in ctg)
ctg = set('American' if 'American' in i else i for i in ctg)

dict_ctg = {}
for i in list(ctg):
    dict_ctg[i] = len(business[business.business_categories.str.contains(i)])
    
df_ctg = pd.DataFrame.from_dict(dict_ctg, orient='index')
df_ctg.sort_values(0, ascending=False, inplace=True)
df_ctg.columns=['count_of_restaurant_type']
exclude = np.asarray(['Restaurants','Food','Fast Food']) # exclude some broad terms
df_ctg = df_ctg[(~df_ctg.index.isin(exclude))]
df_ctg = df_ctg[df_ctg.count_of_restaurant_type>100] # picked the top 7 types

# -add restaurants categories as features
for i in df_ctg.index:
    business[i] = False
    business.loc[business.business_categories.str.contains(i), i] = True

# -categories vs stars and reviews
df_ctg['avg_reviews']=0
df_ctg['avg_stars']=0
for i in df_ctg.index:
    df_ctg.avg_reviews.ix[i] = business[business[i] == True]['business_review_count'].mean()
    df_ctg.avg_stars.ix[i] = business[business[i] == True]['business_stars'].mean()
df_ctg
```

## - What leads to higher review stars?
- loud (-) & WheelcharAccessible (+)
- Burgers places get lower stars

```python
# -multiple regression
features = 'C(business_NoiseLevel)+ C(business_WheelchairAccessible)'
est = smf.ols(formula='business_stars ~'+features, data=business).fit()
est.summary().tables[1]
```

## - Target restaurants for sales
- 16% of restaurants changed at least half a star from 2013 to 2014

```python
# -restaurants with significant drop in reviews/stars (retro at 2014-12-31)
target = pd.DataFrame()
target['past_mh'] = review[review.month=='2014-12'].groupby(['business_id']).size()
target['mh_before'] = review[review.month=='2014-11'].groupby(['business_id']).size()
target['drop'] = 1-target.past_mh/target.mh_before
target = target.sort_values(by=['drop'], ascending=False)
pp(target.head(10)).as_percent('drop')
```

```python
# ratio of restaurants having changed stars in the past year
stars_change = pd.DataFrame()
stars_change['stars_2014'] = review[review.year=='2014'].groupby(['business_id'])['review_stars'].mean()
stars_change['reviews_2014'] = review[review.year=='2014'].groupby(['business_id']).size()
stars_change['stars_2013'] = review[review.year=='2013'].groupby(['business_id'])['review_stars'].mean()
stars_change['reviews_2013'] = review[review.year=='2013'].groupby(['business_id']).size()
stars_change['abs_change'] = abs(stars_change.stars_2014-stars_change.stars_2013)
res_filtered = stars_change[(stars_change.reviews_2014>10)&(stars_change.reviews_2013>10)] # filter out restaurants with very few reviews
r = 1.*len(res_filtered[res_filtered.abs_change>=.5])/len(res_filtered)
print 'Based on {} restaurants with at least 10 reviews each year, {:.1%} have a change of at least half a star from 2013 to 2014.'.format(len(res_filtered),r) 
```

## - User analysis
- who are valuable users (quality users)?
- The elite/quality ratio of new Yelpers each year are constantly dropping (from 60% to 6%)
    - it may take time to become elite or gain popularity in the community
    - but it could possible that users joined in recent years write less high quality reviews

```python
# -measure valuable users - quality
reviewer['elite_ever'] = 0
reviewer.loc[(reviewer.reviewer_years_elite+reviewer.reviewer_currently_elite)>0, 'elite_ever'] = 1
reviewer['quality'] = reviewer.reviewer_num_compliments/reviewer.reviewer_review_count
est = smf.ols(formula='quality ~ reviewer_num_friends+elite_ever', data=reviewer).fit()
est.summary().tables[1]
```

```python
# -quality users/elite over time
usr_quality = reviewer[reviewer.quality>0.16]
usr_elite = reviewer[reviewer.elite_ever>0]
def yr_count(df):
    yr = df.reviewer_yelper_since.str[:4]
    return df[(yr<>'2004')&(yr<>'2015')].reviewer_yelper_since.str[:4].value_counts()
usr = pd.DataFrame({'quality':yr_count(usr_quality), 'elite':yr_count(usr_elite), 'all':yr_count(reviewer)})
usr['pct_elite'] = usr.elite/usr['all']
usr['pct_quality'] = usr.quality/usr['all']
usr[['pct_elite', 'pct_quality']].sort_index().plot(kind='bar', title = 'Quality and Elite Users')
#usr.quality.pct_change()[usr.index>'2011'].mean()
```
